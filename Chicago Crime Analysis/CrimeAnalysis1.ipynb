```python
import datetime
import pandas as pd


def NumberOfCrimeType(typeOfCrime):
    df1 = pd.read_csv("Crimes.csv", header=None, usecols=[2,5])

    jan = 0
    feb = 0
    mar = 0
    apr = 0
    may = 0
    jun = 0
    jul = 0
    aug = 0
    sep = 0
    octo = 0
    nov = 0
    dec = 0
    cnt = 0
    
    for e,f in zip(df1[5],df1[2]):
        if e == typeOfCrime:
            cnt = cnt + 1
            pos = f.find(" ")
            dt = f[:pos]
            l = dt.split("/")
            month = int(l[0])
            if month == 1:
                jan = jan + 1
            elif month == 2:
                feb = feb + 1
            elif month == 3:
                mar = mar + 1
            elif month == 4:
                apr = apr + 1
            elif month == 5:
                may = may + 1
            elif month == 6:
                jun = jun + 1
            elif month == 7:
                jul = jul + 1
            elif month == 8:
                aug = aug + 1
            elif month == 9:
                sep = sep + 1
            elif month == 10:
                octo = octo + 1
            elif month == 11:
                nov = nov + 1
            else:
                dec = dec + 1

    print(typeOfCrime)
    print("Jan: " , jan)
    print("Feb: " , feb)
    print("Mar: " , mar)
    print("Apr: " , apr)
    print("May: " , may)
    print("Jun: " , jun)
    print("Jul: " , jul)
    print("Aug: " , aug)
    print("Sep: " , sep)
    print("Oct: " , octo)
    print("Nov: " , nov)
    print("Dec: " , dec)
    print("Total: ", cnt)


NumberOfCrimeType("ROBBERY")
```
